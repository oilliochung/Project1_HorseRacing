```python
for i in onlybrandlist[828:]:
    driver.find_element_by_name('BrandNumber').clear()
    inputElement = driver.find_element_by_name("BrandNumber")
    inputElement.send_keys(i)
    driver.find_element_by_id('submit1').click()
    
    subhtml2=driver.page_source
    soup2 = BeautifulSoup(subhtml2, "html.parser")
    wait=WebDriverWait(driver,15) 
    
    table = soup2.find_all(class_="horseProfile")
    table_data = [i.find_all('td') for i in table]
    l = "".join([i.text.strip() for i in table_data[0]])
    
    try:
        name = re.findall("(.+)\(\w\d{3}\)", l)[0]
    except:
        name=np.nan
    try:
        country = re.findall("Origin / Age\:(\w+) / \d+", l)[0]
    except:
        try:
            country = re.findall("Country of Origin\\n\:\\n(\w+)", l)[0]
        except:
            country = np.nan
    try:
        age = re.findall("Origin / Age\:\w+ / (\d+)", l)[0]
    except:
        age=np.nan
    try:
        colour= re.findall("Colour / Sex\:(\w+) / \w+Import", l)[0]
    except:
        colour=np.nan
    try:
        sex= re.findall("Colour / Sex\:\w+ / (\w+)Import", l)[0]
    except:
        sex= np.nan
    try:
        importtype=re.findall("Import Type\:(\w+)Season", l)[0]
    except:
        try:
            importtype = re.findall("Import Type\\n\:\\n(\w+)", l)[0]
        except:
            importtype = np.nan
    try:
        seasonstakes = re.findall("Season Stakes\*\:(\$\d+\,?\d*\,?\d*)", l)[0]
    except:
        seasonstakes =np.nan
    try:
        totalstakes = re.findall("Total Stakes\*\:(\$\d+\,?\d*\,?\d*)", l)[0]
    except:
        totalstakes = np.nan
    try:
        starts = re.findall("No\. of 1\-2\-3\-Starts\*\:(\d\-\d\-\d\-\d)", l)[0]
    except:
        starts=np.nan
    try:
        paststarts=re.findall("No\. of starts in past 10race meetings\:(\d+)", l)[0]
    except:
        partstarts =np.nan
    try:
        location = re.findall("Current Stable Location\(Arrival Date\)\:(.+)\(", l)[0]
    except:
        location=np.nan
    try:
        arrival = re.findall("Current Stable Location\(Arrival Date\)\:.+\((\d+/\d+/\d+)\)", l)[0]
    except:
        arrival=np.nan
    try:
        trainer = re.findall("Trainer\:(.+?)Owner", l)[0]
    except:
        trainer=np.nan
    try:
        owner = re.findall("Owner\:(.+?)Current",l)[0]
    except:
        try:
            owner = re.findall("Owner\\n\:\\n(.+?)\\n",l)[0]
        except:
            owner=np.nan
    try:
        currentrating = re.findall("Current Rating\:(\d+)",l)[0]
    except:
        currentrating=np.nan
    try:
        startrating = re.findall("Start ofSeason Rating\:(\d+)",l)[0]
    except:
        startrating=np.nan
    try:
        sire = re.findall("Sire\:(.+?)Dam",l)[0]
    except:
        sire=np.nan
    try:
        dam = re.findall("Dam\:(.+?)Dam",l)[0]
    except:
        dam=np.nan
    try:
        damsire= re.findall("Dam's Sire\:(.+?)Same",l)[0]
    except:
        damsire=np.nan



    d= [[name,i,country,age,colour,sex,importtype,seasonstakes,totalstakes,
     starts,paststarts,location,arrival,trainer,owner,currentrating,
    startrating,sire,dam,damsire]]
    
    profile=pd.DataFrame(d)


    profiles = pd.concat([profiles,profile])
    
    driver.back()
    
    
    
    
    
    
    
    
```
