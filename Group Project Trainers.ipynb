```python
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
import pandas as pd
import numpy as np
import re
```

```python
URL = "https://racing.hkjc.com/racing/information/English/Trainers/TrainerRanking.aspx?Season=Previous&View=Numbers&Racecourse=ALL"
driver = webdriver.Chrome(executable_path='/Applications/chromedriver')
driver.get(URL)
#df=pd.DataFrame()

wait=WebDriverWait(driver,10) #sometimes takes time for the page to load

subhtml = driver.page_source
soup4 = BeautifulSoup(subhtml, "html.parser")
```

```python
table = soup4.find_all(class_="table_bd")
table_data = [i.find_all('td') for i in table]
l = [i.text.strip() for i in table_data[0]]
```

```python
trainers=[]
for i in range(int(len(l[2:])/8)):
    trainers.append(l[2+8*i:2+8*i+8])
    print(trainers)
        
```

```python
trainerframe=pd.DataFrame(trainers)
```

```python
trainerframe.columns = trainerframe.iloc[0]
trainerframe=trainerframe.drop(0)
trainerframe
```

```python
trainerframe.to_csv('trainers.csv',index=False)
```
