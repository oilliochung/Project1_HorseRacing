```python
import pandas
jockeys=pandas.read_csv('./data/jockeys.csv')
jockeyspercent=pandas.read_csv(r'./data/jockeys_pct.csv')
trainers=pandas.read_csv('./data/trainers.csv')
trainerspercent=pandas.read_csv(r'./data/trainers_pct.csv')
horsesprofiles=pandas.read_csv('./data/horses.csv')
main=pandas.read_csv('./data/full_season_performance_clean.csv')
pd.options.display.max_columns = None   #display all columns in jupyter
horsesprofiles.columns=['H_id', 'H_name', 'H_country', 'H_age', 'H_colour', 'H_sex', 'H_ImportType', 'H_season_stakes', 'H_ttl_stakes', 'H_no_of_1_2_3_Starts', 'H_no_of_starts_in_past10', 'H_curr_location', 'H_arr_date', 'H_trainer', 'H_owner', 'H_curr_rating', 'H_1920rating', 'H_sire', 'H_dam', 'H_damSire']
#main['horse_id']=main['horse'].apply(lambda x: x[-5:-1])

```

```python
main1=ps.sqldf('''
select a.* , --main table

--add Jockey Infor
b.Win as J_Win, c.Win as J_Win_perc, 
b.'2nd' as J_2nd,  c.'2nd' as J_2nd_perc,
b.'3rd' as J_3rd,  c.'3rd' as J_3rd_perc,
b.'4th' as J_4th,  c.'4th' as J_4th_perc,
b.'5th' as J_5th,  c.'5th' as J_5th_perc,

--add Trainers Infor
b.'Total Rides' as J_ttl_rides, c.'Stakes Won' as J_stakes_won,
d.Win as T_Win, e.Win as T_Win_perc,
d.'2nd' as T_2nd,  e.'2nd' as T_2nd_perc,
d.'3rd' as T_3rd,  e.'3rd' as T_3rd_perc,
d.'4th' as T_4th,  e.'4th' as T_4th_perc,
d.'5th' as T_5th,  e.'5th' as T_5th_perc,
d.'Total Runs' as T_ttl_runs, e.'Stakes Won' as T_stakes_won,

--add Horse Infor
f.*

from main a 
left join jockeys b on a.jockey=b.Jockey
left join jockeyspercent c on a.jockey=c.Jockey
left join trainers d on a.trainer=d.Trainer
left join trainerspercent e on a.trainer=e.Trainer
left join horsesprofiles f on a.horse_id=f.H_id
''')


```

```python
#clean str other than no.
main1['J_Win_perc']=main1['J_Win_perc'].str.strip('%')
main1['J_2nd_perc']=main1['J_2nd_perc'].str.strip('%')
main1['J_3rd_perc']=main1['J_3rd_perc'].str.strip('%')
main1['J_4th_perc']=main1['J_4th_perc'].str.strip('%')
main1['J_5th_perc']=main1['J_5th_perc'].str.strip('%')
main1['J_stakes_won']=main1['J_stakes_won'].str.strip('$')

main1['T_Win_perc']=main1['T_Win_perc'].str.strip('%')
main1['T_2nd_perc']=main1['T_2nd_perc'].str.strip('%')
main1['T_3rd_perc']=main1['T_3rd_perc'].str.strip('%')
main1['T_4th_perc']=main1['T_4th_perc'].str.strip('%')
main1['T_5th_perc']=main1['T_5th_perc'].str.strip('%')
main1['T_stakes_won']=main1['T_stakes_won'].str.strip('$')

#main1['H_season_stakes']=main1['H_season_stakes'].str.strip('$')
#main1['H_ttl_stakes']=main1['H_ttl_stakes'].str.strip('$')
```
